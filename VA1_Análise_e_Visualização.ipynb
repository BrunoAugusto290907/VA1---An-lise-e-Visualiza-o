```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://dados.recife.pe.gov.br/dataset/0384775d-d187-4cf7-aa64-102c95c139de/resource/326b7fb6-90af-46d9-b342-a9c8fa6a706c/download/matriculas-na-rede-municipal-do-ano-de-2024.csv"

df = pd.read_csv(url, sep=';', encoding='utf-8')

df.columns = df.columns.str.lower()
```

1. Qual é o total de matrículas na base de dados?

```python
total_matriculas = df.shape[0]
print(f"Total de matrículas: {total_matriculas}")
```

2. Quantos alunos são do sexo feminino e quantos são do sexo masculino?

```python
contagem_sexo = df['sexo'].value_counts()
print(contagem_sexo)
```

3. Qual é a idade média dos alunos?

```python
idade_media = df['idade'].mean()
print(f"Idade Média: {idade_media:.2f} anos")
```

4. Qual é a idade mínima e máxima dos alunos?

```python
idade_min = df['idade'].min()
idade_max = df['idade'].max()
print(f"Idade Mínima: {idade_min}")
print(f"Idade Máxima: {idade_max}")
```

5. Quantidade de alunos por turno

```python
alunos_turno = df['turno'].value_counts()
print(alunos_turno)
```

6. Quantidade de alunos por modalidade de ensino.

```python
alunos_modalidade = df['modalidade'].value_counts()
print(alunos_modalidade)
```

7. Qual turma possui mais alunos?

```python
turma_mais_alunos = df['turma'].value_counts().idxmax()
qnt_alunos_turma = df['turma'].value_counts().max()
print(f"A turma com mais alunos é a '{turma_mais_alunos}' com {qnt_alunos_turma} alunos.")
```

1. Distribuição por sexo

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='sexo', palette='pastel')
plt.title('Distribuição de Alunos por Sexo')
plt.ylabel('Quantidade de Alunos')
plt.xlabel('Sexo')
plt.show()
```

2. Distribuição por turno

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='turno', palette='viridis')
plt.title('Distribuição de Alunos por Turno')
plt.ylabel('Quantidade de Alunos')
plt.xlabel('Turno')
plt.show()
```

3. Distribuição de idades

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='idade', bins=20, kde=True, color='skyblue')
plt.title('Histograma da Distribuição de Idades')
plt.ylabel('Frequência')
plt.xlabel('Idade')
plt.show()
```

4. Idade por sexo

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='sexo', y='idade', palette='Set2')
plt.title('Distribuição de Idade por Sexo')
plt.ylabel('Idade')
plt.xlabel('Sexo')
plt.show()
```

5. Alunos por turma

```python
plt.figure(figsize=(12, 6))
top_turmas = df['turma'].value_counts().head(15)

sns.barplot(x=top_turmas.index, y=top_turmas.values, palette='mako')
plt.title('Top 15 Turmas com Mais Alunos')
plt.ylabel('Quantidade de Alunos')
plt.xlabel('Turma')
plt.xticks(rotation=45, ha='right')
plt.show()
```

6. Idade média por turno

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='turno', y='idade', errorbar=None, palette='rocket')
plt.title('Idade Média dos Alunos por Turno')
plt.ylabel('Idade Média (Anos)')
plt.xlabel('Turno')
plt.show()
```
